### C. Normalize data for comparative analysis

### apply Min-Max normalization to the ConvertedCompYearly column, creating a new column ConvertedCompYearly_MinMax with normalized values.

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
scaler = MinMaxScaler()
df_cleaned = df.dropna(subset=['ConvertedCompYearly']).copy()
df_cleaned['ConvertedCompYearly_MinMax'] = scaler.fit_transform(df_cleaned[['ConvertedCompYearly']])
print(df_cleaned[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax']].head())
print(f"\nMinimum value: {df_cleaned['ConvertedCompYearly_MinMax'].min()}")
print(f"Maximum value: {df_cleaned['ConvertedCompYearly_MinMax'].max()}")
```

### calculate Z-scores for the ConvertedCompYearly column, saving the results in a new column ConvertedCompYearly_Zscore.

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
df_cleaned = df.dropna(subset=['ConvertedCompYearly']).copy()
df_cleaned['ConvertedCompYearly_Zscore'] = std_scaler.fit_transform(df_cleaned[['ConvertedCompYearly']])
print(df_cleaned[['ConvertedCompYearly', 'ConvertedCompYearly_Zscore']].head())
print(f"\nNew Mean (المتوسط الجديد): {df_cleaned['ConvertedCompYearly_Zscore'].mean():.2f}")
print(f"New Std Dev (الانحراف المعياري الجديد): {df_cleaned['ConvertedCompYearly_Zscore'].std():.2f}")
```

### D. Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.

```python
country_counts = df['Country'].value_counts()
print("Top 20 countries:")
print(country_counts.head(20))
unique_countries = df['Country'].unique()
print(f"\nTotal unique entries in Country: {len(unique_countries)}")
```

```python
min_respondents = 50
countries_to_keep = country_counts[country_counts > min_respondents].index

df_filtered = df[df['Country'].isin(countries_to_keep)]

print(f"Data shape before filtering: {df.shape}")
print(f"Data shape after filtering: {df_filtered.shape}")
```

### Standardization

```python
country_mapping = {
    'United States of America': 'United States',
    'USA': 'United States',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'UK': 'United Kingdom'
}

df['Country'] = df['Country'].replace(country_mapping)
```

```python
print("Original Education Levels:")
print(df['EdLevel'].unique())

ed_mapping = {
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor',
    'Master’s degree (M.A., M.S., M.Eng., etc.)': 'Master',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'High School',
    'Some college/university study without earning a degree': 'Undergrad',
    'Professional degree (JD, MD, etc.)': 'Professional',
    'Other doctoral degree (Ph.D., Ed.D., etc.)': 'PhD',
    'Associate degree (A.A., A.S., etc.)': 'Associate',
    'Something else': 'Other',
    'Primary/elementary school': 'Primary School'
}

df['EdLevel'] = df['EdLevel'].replace(ed_mapping)

print("\nStandardized Education Levels:")
print(df['EdLevel'].value_counts())
```
